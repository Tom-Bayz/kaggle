```python
cum = train_df.groupby(by='user_id')['content_type_id'].agg(['cumsum', 'cumcount'])
train_df['user_lecture_cumsum'] = cum['cumsum'] 
train_df['user_lecture_lv'] = cum['cumsum'] / cum['cumcount']
```

```python
train_df.plot.scatter(x='user_lecture_cumsum',
                      y="answered_correctly")
plt.grid(True)
plt.show()
```

```python
lv = train_df.groupby(by='user_id')[['content_type_id']].mean()
ratio = train_df[train_df["answered_correctly"] != -1].groupby(by='user_id')[['answered_correctly']].mean()

(lv.join(ratio)).plot.scatter(x='content_type_id',
                              y="answered_correctly",
                              alpha=0.1)
plt.grid(True)
plt.show()
```
