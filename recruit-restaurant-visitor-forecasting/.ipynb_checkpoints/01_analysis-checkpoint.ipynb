# Data Analysis

```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt

data = {}
data["hpg"] = pd.read_csv("hpg.csv",index_col = 0)
data["air"] = pd.read_csv("air.csv",index_col = 0)
```

```python
data["air"]["reserve_visitors"].plot()
plt.grid(True)
plt.show()


plt.hist(data["air"]["reserve_visitors"],bins=30)
plt.grid(True)
```
