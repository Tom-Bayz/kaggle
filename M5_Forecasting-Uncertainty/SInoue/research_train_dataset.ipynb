```python
# Read in the data
INPUT_DIR = '/kaggle/input/m5-forecasting-uncertainty/'
cal = pd.read_csv(f'{INPUT_DIR}/calendar.csv')
stv = pd.read_csv(f'{INPUT_DIR}/sales_train_validation.csv')
ss = pd.read_csv(f'{INPUT_DIR}/sample_submission.csv')
sellp = pd.read_csv(f'{INPUT_DIR}/sell_prices.csv')
```

```python
l1 = []
foo_l2 = []
hob_l2 = []
hou_l2 = []
foo_l3 = [[],[],[]]
hob_l3 = [[],[]]
hou_l3 = [[],[]]
for s in stv['item_id']:
    a,b,c = s.split('_')
    l1.append(a)
    if a == 'FOODS':
        foo_l2.append(b)
        if b == '1':
            foo_l3[0].append(c)
        elif b == '2':
            foo_l3[1].append(c)
        elif b == '3':
            foo_l3[2].append(c)
    elif a == 'HOBBIES':
        hob_l2.append(b)
        if b == '1':
            hob_l3[0].append(c)
        elif b == '2':
            hob_l3[1].append(c)
    elif a == 'HOUSEHOLD':
        hou_l2.append(b)
        if b == '1':
            hou_l3[0].append(c)
        elif b == '2':
            hou_l3[1].append(c)
    else:
        print(s)
```
